```python
# 利用CNN替代clip_by_global_norm来解决梯度爆炸
# idea: 利用卷积核提取特征（将卷积核的通道数压缩成8个），从而使CNN只需要记忆少量的信息，从而不会出现断崖式梯度
#       利用卷积核将词压缩成词组，并提取主要词组
```

```python
params = {
  'vocab_path': './vocab/word.txt',
  'train_path': './data/train.txt',
  'test_path': './data/test.txt',
  'embedding_path': './data/glove.6B.50d.txt',
  'num_samples': 25000,
  'num_labels': 2,
  'batch_size': 32,
  'max_len': 1000,
  'rnn_units': 200,
  'dropout_rate': 0.2,
  'clip_norm': 10.,
  'num_patience': 3,
  'lr': 3e-4,
}
```

```python
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np

import tensorflow as tf

from tqdm import tqdm
import logging

import time
```

# 1. 获取word2idx映射表（语料表）

```python
word2idx = {}


with open(params['vocab_path'], 'r',encoding='utf-8') as f:
    for i, w in enumerate(f.readlines()):
        w = w.strip()
        word2idx[w] = i
    
```

```python
# + 1 表示如果不在语料表中，就都是unknow，值为[0, 0, 0]
embedding = np.zeros((len(word2idx)+1, 50))

with open(params['embedding_path'], 'r', encoding='utf-8') as f:
    count = 0
    lines = f.readlines()
    
    batch = len(lines) // 5
    progress_bar = tqdm(total=batch, desc="read data")
    
    for i, line in enumerate(lines):
        line = line.strip()
        line = line.split(' ')
        word, vec = line[0], line[1:]

        if word in word2idx.keys():
            embedding[word2idx[word]] = np.asarray(vec, dtype='float32')
            count += 1
            

        if (i % 5 == 0):
            progress_bar.update(5)

    progress_bar.close()
    
```

```python
def data_generator(path, params):
    with open(path, 'r', encoding='utf-8') as f:
        for line in f.readlines():
            line = line.strip()
            label, content = line.split('\t')
            content = content.split(' ')
            content_idx = [word2idx.get(w, len(word2idx)) for w in content]
            if len(content_idx) >= params['max_len']:                 #截断操作
                content_idx = content_idx[:params['max_len']]
            else:
                content_idx += [0] * (params['max_len'] - len(content_idx))     #补齐操作

            yield content_idx, int(label)
```

```python
def dataset(is_training, params):
    _shapes = ([params['max_len']], ())
    _types = (tf.int32, tf.int32)
  
    if is_training:
        ds = tf.data.Dataset.from_generator(
            lambda: data_generator(params['train_path'], params),
            output_shapes = _shapes,
            output_types = _types,)
        ds = ds.shuffle(params['num_samples'])
        ds = ds.batch(params['batch_size'])
        ds = ds.prefetch(tf.data.experimental.AUTOTUNE)#设置缓存序列，根据可用的CPU动态设置并行调用的数量，说白了就是加速
    else:
        ds = tf.data.Dataset.from_generator(
            lambda: data_generator(params['test_path'], params),
            output_shapes = _shapes,
            output_types = _types,)
        ds = ds.batch(params['batch_size'])
        ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
  
    return ds
```

```python
def is_descending(history: list):
    history = history[-(params['num_patience']+1):]
    for i in range(1, len(history)):
        if history[i-1] <= history[i]:
            return False
    return True  
```

```python
class CustomModel(tf.keras.Model):
    def __init__(self, params):
        super(CustomModel, self).__init__()
        
        self.embedding = tf.Variable(embedding, dtype=tf.float32,
                                                name='pretrained_embedding',
                                                trainable=False)
        self.conv = tf.keras.layers.Conv2D(filters=8, kernel_size=(5, self.embedding.shape[1]), strides=1,
                                          padding='valid')
        
        self.max_pool = tf.keras.layers.MaxPooling2D(pool_size=(4, 1), padding='valid')
        self.connect = tf.keras.layers.Concatenate(axis=1)
        
        self.rnn1 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(params['rnn_units'], return_sequences=True))
        self.rnn2 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(params['rnn_units'], return_sequences=True))
        self.rnn3 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(params['rnn_units'], return_sequences=False))
        
        self.fc = tf.keras.layers.Dense(2*params['rnn_units'], tf.nn.elu)
        self.out_linear = tf.keras.layers.Dense(2)
        
        self.drop0 = tf.keras.layers.Dropout(params['dropout_rate'])
        self.drop1 = tf.keras.layers.Dropout(params['dropout_rate'])
        self.drop2 = tf.keras.layers.Dropout(params['dropout_rate'])
        self.drop3 = tf.keras.layers.Dropout(params['dropout_rate'])

    
    
    def call(self, inputs, training=False):
        embd = tf.nn.embedding_lookup(self.embedding, inputs)
        embd = tf.expand_dims(embd, axis=-1)
        
        x = self.conv(embd)
        x = self.max_pool(x)
        self.drop0(x, training=training)
        
        x = tf.reshape(x, (x.shape[0], x.shape[1], x.shape[3]))
        x = tf.transpose(x, perm=[0, 2, 1])
        
        x = self.rnn1(x)
        self.drop1(x, training=training)
        x = self.rnn2(x)
        self.drop2(x, training=training)
        x = self.rnn3(x)
        self.drop3(x, training=training)
        
        x = self.fc(x)
        
        x = self.out_linear(x)
        
        return x
```

```python
word2idx = {}
with open(params['vocab_path'],encoding='utf-8') as f:
    for i, line in enumerate(f):
        line = line.rstrip()
        word2idx[line] = i
params['word2idx'] = word2idx
params['vocab_size'] = len(word2idx) + 1


model = CustomModel(params)
# model.build(input_shape=(None, None))#设置输入的大小，或者fit时候也能自动找到
#pprint.pprint([(v.name, v.shape) for v in model.trainable_variables])

#链接：https://tensorflow.google.cn/api_docs/python/tf/keras/optimizers/schedules/ExponentialDecay?version=stable
#return initial_learning_rate * decay_rate ^ (step / decay_steps)
decay_lr = tf.optimizers.schedules.ExponentialDecay(params['lr'], 1000, 0.95)#相当于加了一个指数衰减函数
optim = tf.optimizers.Adam(params['lr'])
global_step = 0

history_acc = []
best_acc = .0

t0 = time.time()
logger = logging.getLogger('tensorflow')
logger.setLevel(logging.INFO)
```

```python
losses_list = []

for i in range(20):
  # 训练模型
    for texts, labels in dataset(is_training=True, params=params):
        with tf.GradientTape() as tape:#梯度带，记录所有在上下文中的操作，并且通过调用.gradient()获得任何上下文中计算得出的张量的梯度
            logits = model(texts, training=True)
            loss = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=labels, logits=logits)
            loss = tf.reduce_mean(loss)
  
        losses_list.append(loss.numpy().item())
        optim.lr.assign(decay_lr(global_step))
        grads = tape.gradient(loss, model.trainable_variables)
        # grads, _ = tf.clip_by_global_norm(grads, params['clip_norm'])
        optim.apply_gradients(zip(grads, model.trainable_variables))#更新梯度

        if global_step % 50 == 0:
            logger.info("Step {} | Loss: {:.4f} | Spent: {:.1f} secs | LR: {:.6f}".format(
                global_step, loss.numpy().item(), time.time()-t0, optim.lr.numpy().item()))
            t0 = time.time()
            
            
        global_step += 1
        
    

    # 验证集效果
    m = tf.keras.metrics.Accuracy()

    for texts, labels in dataset(is_training=False, params=params):
        logits = model(texts, training=False)
        y_pred = tf.argmax(logits, axis=-1)
        m.update_state(y_true=labels, y_pred=y_pred)
    
    acc = m.result().numpy()
    logger.info("Evaluation: Testing Accuracy: {:.3f}".format(acc))
    history_acc.append(acc)
  
    if acc > best_acc:
        best_acc = acc
    logger.info("Best Accuracy: {:.3f}".format(best_acc))
  
    if len(history_acc) > params['num_patience'] and is_descending(history_acc):
        logger.info("Testing Accuracy not improved over {} epochs, Early Stop".format(params['num_patience']))
        break
```

```python
import pandas as pd

loss = 0
l_l = []
for i, item in enumerate(losses_list):
    if i % 15 == 0:
        loss /= 16
        l_l.append(loss)
        loss = 0
    
    loss += item
    
pd.Series(l_l[1:]).plot(figsize=(15, 8))
```
